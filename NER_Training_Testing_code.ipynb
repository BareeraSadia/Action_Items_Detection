```python
import json
import spacy
from spacy.training import Example
import random
```

```python

with open("final data/train_annotations.json", "r", encoding="utf-8") as f:
    data = json.load(f)

classes = data['classes']
annotations = data['annotations']
```

```python

training_data = []
for entry in annotations:
    text = entry[0]
    entities = entry[1]['entities']
    training_data.append((text, {"entities": entities}))
```

```python

nlp = spacy.blank("xx")  
ner = nlp.add_pipe("ner")
for label in classes:
    ner.add_label(label) 
```

```python

optimizer = nlp.begin_training()
for epoch in range(50): 
    random.shuffle(training_data)
    for text, annotations in training_data:
        doc = nlp.make_doc(text)
        example = Example.from_dict(doc, annotations)
        nlp.update([example], drop=0.5, sgd=optimizer)
        print(epoch)
```

```python
import os
output_dir = "ner_model"  
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

nlp.to_disk(output_dir)
print(f"Model saved to {output_dir}")
```

```python
import spacy
import json
from spacy.training import Example


model_path = "ner_model"  
nlp = spacy.load(model_path)
test_file_path = "jsons/test_annotations.json"
with open(test_file_path, "r", encoding="utf-8") as f:
    test_data = json.load(f)
test_annotations = test_data['annotations']
total_entities = 0
correct_entities = 0
detailed_results = []
for annotation in test_annotations:
    text = annotation[0]  
    true_entities = annotation[1]['entities']  
    doc = nlp(text)
    predicted_entities = [(ent.start_char, ent.end_char, ent.label_) for ent in doc.ents]
    total_entities += len(true_entities)
    for true_entity in true_entities:
        if true_entity in predicted_entities:
            correct_entities += 1

    detailed_results.append({
        "text": text,
        "true_entities": true_entities,
        "predicted_entities": predicted_entities
    })

output_evaluation_path = "test_evaluation_results.json"
with open(output_evaluation_path, "w", encoding="utf-8") as out_file:
    json.dump(detailed_results, out_file, ensure_ascii=False, indent=4)
print(f"Detailed results saved to {output_evaluation_path}")
```

```python
import json
from collections import defaultdict

def calculate_overlap(true_entity, predicted_entity, threshold=0.8):
    """
    Calculate the overlap between true and predicted entities based on IoU.
    
    :param true_entity: Tuple of (start, end, label) for true entity.
    :param predicted_entity: Tuple of (start, end, label) for predicted entity.
    :param threshold: Overlap threshold for considering a match.
    :return: Boolean indicating if the entities match based on overlap.
    """
    true_start, true_end, true_label = true_entity
    pred_start, pred_end, pred_label = predicted_entity
    overlap_start = max(true_start, pred_start)
    overlap_end = min(true_end, pred_end)
    
    overlap = max(0, overlap_end - overlap_start)  
    true_length = true_end - true_start  
    predicted_length = pred_end - pred_start 
    overlap_ratio = overlap / float(true_length + predicted_length - overlap)
    
    return overlap_ratio >= threshold
with open("test_evaluation_results.json", "r", encoding="utf-8") as f:
    eval_results = json.load(f)
total_entries = len(eval_results)
true_positive = 0
false_positive = 0
false_negative = 0
label_metrics = defaultdict(lambda: {"tp": 0, "fp": 0, "fn": 0})
for entry in eval_results:
    text = entry["text"]
    true_entities = entry["true_entities"]
    predicted_entities = entry["predicted_entities"]
    
    for true_entity in true_entities:
        matched = False
        true_start, true_end, true_label = true_entity
        for pred_entity in predicted_entities:
            if calculate_overlap(true_entity, pred_entity):  
                true_positive += 1
                label_metrics[true_label]["tp"] += 1
                matched = True
                break
        if not matched: 
            false_negative += 1
            label_metrics[true_label]["fn"] += 1
    
    for pred_entity in predicted_entities:
        matched = False
        pred_start, pred_end, pred_label = pred_entity
        for true_entity in true_entities:
            if calculate_overlap(true_entity, pred_entity):  
                matched = True
                break
        if not matched: 
            false_positive += 1
            label_metrics[pred_label]["fp"] += 1
precision = true_positive / (true_positive + false_positive) if true_positive + false_positive > 0 else 0
recall = true_positive / (true_positive + false_negative) if true_positive + false_negative > 0 else 0
f1_score = 2 * (precision * recall) / (precision + recall) if precision + recall > 0 else 0
print(f"Total entries in evaluation results: {total_entries}")
print(f"Overall Precision: {precision:.4f}")
print(f"Overall Recall: {recall:.4f}")
print(f"Overall F1 Score: {f1_score:.4f}")
print("\nClassification Report for Each Label:")
for label, metrics in label_metrics.items():
    tp, fp, fn = metrics["tp"], metrics["fp"], metrics["fn"]
    precision_label = tp / (tp + fp) if tp + fp > 0 else 0
    recall_label = tp / (tp + fn) if tp + fn > 0 else 0
    f1_label = 2 * (precision_label * recall_label) / (precision_label + recall_label) if precision_label + recall_label > 0 else 0
    
    print(f"\nLabel: {label}")
    print(f"Precision: {precision_label:.4f}")
    print(f"Recall: {recall_label:.4f}")
    print(f"F1 Score: {f1_label:.4f}")
results_summary = {
    "Overall Precision": precision,
    "Overall Recall": recall,
    "Overall F1 Score": f1_score,
    "Label Metrics": label_metrics,
    "Total Entries": total_entries
}
with open("evaluation_summary.json", "w", encoding="utf-8") as output_file:
    json.dump(results_summary, output_file, indent=4)
print("Evaluation summary saved to evaluation_summary.json")
```
